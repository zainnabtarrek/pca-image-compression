***Importing the required libraries and modules***

```python
import kagglehub # kagglehub for downloading the dataset used
import os # os module to navigate through the module
import cv2 # a module to help in calculating the covariance and store image pixels
import numpy as np # a module to help in the calculations with matrices in general
import matplotlib.pyplot as plt # a module to plot the image
from sklearn.preprocessing import StandardScaler # a scaler to standardize the pixels
```

```python
# Download latest version
path = kagglehub.dataset_download("nih-chest-xrays/sample")

print("Path to dataset files:", path)
```

**initializing the shape of the images matrix**

```python
image_shape = (1024, 1024)
num_images = 5606
folder_path = path + "/sample/images"
```

***loading the images from a file***

```python
image_vectors = np.empty((num_images, image_shape[0] * image_shape[1]),
                         dtype=np.uint8)

for i, filename in enumerate(os.listdir(folder_path)):
    if filename.endswith('.png'):
        file_path = os.path.join(folder_path, filename)

        # Load and flatten the image
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        image_vectors[i] = image.flatten()
```

***calculating the covariance for the image pixels as features***

```python
# setting rowvar to False make cov method deal with rows as observations
# and with columns as variables as the regular usage of the method is the opposite
covariance_matrix = np.cov(standardized_image, rowvar=False)
print(covariance_matrix)
print(covariance_matrix.shape)
```

***Sorting the eigen vectors and values descendatly accordingly***

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
print(eigenvalues.shape)
print(eigenvectors.shape)

sorted_indices = np.argsort(eigenvalues)[::-1]

# Step 2: Sort eigenvalues and eigenvectors using the sorted indices
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]
print(sorted_eigenvectors.shape)

# Create a dictionary mapping sorted eigenvalues to their corresponding eigenvectors
eigen_dict = {val: vec for val, vec in zip(sorted_eigenvalues, sorted_eigenvectors.T)}

# Print the dictionary
print(len(eigen_dict))
```

***printing the eigen values to determine how many to use***

```python
for i in range(17):
    print(eigenvalues[sorted_indices[i]])
    # print(eigen_dict[sorted_eigenvalues[i]])
```

***Multiplying the vectors to get the reduced matrix***

```python
new_eigenvectors = sorted_eigenvectors[:, :100]
new_image = np.dot(standardized_image, new_eigenvectors)
print(new_image.shape)
```

***reconstructing the image from the eigen vectors used to reduce it***

```python
reconstructed_image = np.dot(new_image, new_eigenvectors.T)
print(reconstructed_image.shape)
```

***Plotting the recostructed image***

```python
plt.imshow(reconstructed_image, cmap='gray')  # Use grayscale colormap
plt.axis('off')  # Turn off axis
plt.show()
```
